```python
###Q1
import numpy as np
import pandas as pd
import pymc as pm
import arviz as az
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer

# 1. Load dataset
data = load_breast_cancer()
X = data.data
y = data.target  # Binary target variable (0 or 1)

# 2. Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 4. Get dataset dimensions
n, p = X_train.shape

# 5. Define Bayesian Logistic Regression model in PyMC
with pm.Model() as logistic_model:
    # Prior distribution: regression coefficients (betas) follow standard normal distribution
    betas = pm.Normal('betas', mu=0, sigma=1, shape=p)
    intercept = pm.Normal('intercept', mu=0, sigma=1)
    
    # Linear combination
    linear_combination = intercept + pm.math.dot(X_train, betas)
    
    # Logistic regression using logit link function
    probability = pm.Deterministic('p', pm.math.sigmoid(linear_combination))
    
    # Bernoulli likelihood for binary outcome
    y_obs = pm.Bernoulli('y_obs', p=probability, observed=y_train)

    # 6. Perform MCMC sampling
    trace = pm.sample(2000, tune=1000, cores=2, target_accept=0.9)

# 7. MCMC Results: Plot trace and posterior distributions
az.plot_trace(trace, var_names=['betas', 'intercept'])
plt.show()

# 8. MCMC Diagnostics: Gelman-Rubin (R-hat) values should be close to 1
print(az.summary(trace, var_names=['betas', 'intercept']))
```

## **📌 What is Bayesian Multivariate Classification?**
Bayesian multivariate classification models **multiple binary outcome variables** while incorporating **prior knowledge and uncertainty**. This is useful when:
- The outcomes are **not independent**.
- We need **posterior distributions** over regression parameters.
- The dataset is **small**, making **Bayesian inference more robust**.

---

## **📌 How Does the Model Work?**
### **1️⃣ Latent Variable Representation (`z`)**
- Each binary outcome is modeled using a **normally distributed latent variable (`z`)**.
- `z` follows a **Multivariate Normal distribution**.
- The probability of an outcome being **1 or 0** is derived using a **Probit link function**.

### **2️⃣ Bayesian Priors**
- Regression coefficients (`betas`) follow a **Normal prior**.
- The error structure follows a **Multivariate Normal distribution**.

### **3️⃣ Classification using Probit Function**
- We convert `z` into a **probability** using `pm.math.invprobit(z)`.
- This ensures the value lies between **0 and 1**, making it valid for classification.

---

## **📌 Steps in the Model**
### **1️⃣ Generate Synthetic Data**
- Create predictors (`X`).
- Generate latent variables (`z`).
- Convert `z` into binary outcomes (`y`).

### **2️⃣ Define the Bayesian Model**
- Set priors on **regression coefficients (`betas`)**.
- Define **latent normal variables (`z`)**.
- Apply the **Probit transformation** to convert `z` into a probability.

### **3️⃣ Perform MCMC Sampling**
- Use **Markov Chain Monte Carlo (MCMC)** to estimate posterior distributions.

### **4️⃣ Check Convergence & Interpret Results**
- Examine **trace plots**.
- Evaluate **posterior distributions** of regression coefficients.
- Check **Effective Sample Size (ESS)** and **R-hat** for model convergence.

### **3️⃣ Effective Sample Size (ESS) & R-hat**

- **R-hat should be ~1** (indicating convergence).
